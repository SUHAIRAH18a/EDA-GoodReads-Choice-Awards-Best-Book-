## Exploratory Data Analysis Project: Used Cars Analysis Case Study

**OBJECTIVE** : To analyze the used car's price and how EDA focuses on identifying the factors influencing the car price.

**DESCRIPTION** : The project focused on analyzing a dataset of used cars. The dataset included car's brand, model, kilometers driven, price and others relevant information. The project involved loading the data, cleaning and preprocessing it, handling missing values and performing exploratory data analysis (EDA).

**SKILLS** : Data cleaning, Feature engineering, data visualization.

**TECHNOLOGY** : Python, Pandas, Numpy, Seaborn, Matplotlib.

**RESULTS** : Using python functions the analysis revealed Price have high correlations between Engine as well Power.


**<h5>1. Data Collecting**

```python
#import libraries
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 

#to ignore warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
#read the dataset
data = pd.read_csv("used_cars_data.csv")

#show the samples of the data
data.sample(20)
```

**<h5>2. Data Processing** 

**A. Data Reduction**

```python
# remove the unused column are S.No 
data.drop(['S.No.'], axis=1, inplace=True)
```

**B. Missing Value Calculation**

```python
#check how many missing value from each columns
data.isnull().sum()
```

```python
#calculate missing value by percent
(data.isnull().sum() / (len(data)))*100
```

**Observation** : The percentage of missing values for the columns New_Price and Price is ~86% and ~17% respectively


**C. Convert the Data Type**

```python
data.info()
```

```python
# Extract the first element from the string 
data[['Mileage', 'Engine', 'Power', 'New_Price']] = data[['Mileage', 'Engine', 'Power', 'New_Price']].apply(lambda x: x.str.split().str.get(0))
```

```python
# Convert columns to appropriate data types and handle 'null' values (numeric columns)
data = data.astype({'Mileage': 'float', 'Engine': 'float', 'New_Price': 'float'})
data['Power'] = data['Power'].replace('null', np.nan).astype(float)
```

```python
# Convert columns to appropriate data types (categorical columns)
data = data.astype({'Name': 'category', 'Location': 'category', 'Fuel_Type': 'category', 'Transmission': 'category', 'Owner_Type': 'category'})
```

```python
#new data type
data.info()
```

```python
#dataframe after convert
data.head()
```

**<h4>3. Feature Engineering**

**A. Create a new column, "Car_Age" to know the age of the car.**

```python
#create a new column for Car_Age
from datetime import date
date.today().year
data['Car_Age']= date.today().year-data['Year']
data.head()
```

**B. Split the name and create new variables "Brand" and "Model"**

```python
#create new column with variable name Brand
data['Brand'] = data.Name.str.split().str.get(0)
```

```python
#create new column with variable name Model
data['Model'] = data.Name.str.split().str.get(1) + data.Name.str.split().str.get(2)
```

```python
#convert the data type for Brand & Model
data = data.astype({'Brand' : 'category', 'Model': 'category'})
```

**<h5>6. Data Transformation**

Price and Kilometer_Driven variables are highly skewed and on a larger scale, need a log transformation. Log transformation can help in normalization.

```python
#Function for log transformation of the column
def log_transform(data,col):
    for colname in col:
        if (data[colname] == 1.0).all():
            data[colname + '_log'] = np.log(data[colname]+1)
        else:
            data[colname + '_log'] = np.log(data[colname])
    data.info()            
```

```python
log_transform(data,['Kilometers_Driven','Price'])
```

```python
#Log transformation of the feature 'Kilometers_Driven'
sns.distplot(data["Kilometers_Driven_log"], axlabel="Kilometers_Driven_log");
```

```python
#Log transformation of the feature 'Price'
sns.distplot(data["Price_log"], axlabel="Price_log");
```

```python
plt.figure(figsize=(13,17))
sns.pairplot(data=data.drop(['Kilometers_Driven', 'Price'], axis=1))
plt.show()
```

A bar plot can be used to show the relationship between Categorical variables and continuous variables

```python
fig, axarr = plt.subplots(4, 2, figsize=(18, 18))

#First Plot: Location Vs Price
data.groupby('Location')['Price_log'].mean().sort_values(ascending=False).plot.bar(ax=axarr[0][0], fontsize=12)
axarr[0][0].set_title("Location Vs Price", fontsize=18)

#Second Plot: Transmission Vs Price
data.groupby('Transmission')['Price_log'].mean().sort_values(ascending=False).plot.bar(ax=axarr[0][1], fontsize=12)
axarr[0][1].set_title("Transmission Vs Price", fontsize=18)

#Third Plot: Fuel_type Vs Price
data.groupby('Fuel_Type')['Price_log'].mean().sort_values(ascending=False).plot.bar(ax=axarr[1][0], fontsize=12)
axarr[1][0].set_title("Fuel_Type Vs Price", fontsize=18)

#Fourth Plot: Owner_Type Vs Price
data.groupby('Owner_Type')['Price_log'].mean().sort_values(ascending=False).plot.bar(ax=axarr[1][1], fontsize=12)
axarr[1][1].set_title("Owner_Type Vs Price", fontsize=18)

#Fifth Plot: Brand Vs Price
data.groupby('Brand')['Price_log'].mean().sort_values(ascending=False).head(10).plot.bar(ax=axarr[2][0], fontsize=12)
axarr[2][0].set_title("Brand Vs Price", fontsize=18)

#Sixth Plot: Model Vs Price
data.groupby('Model')['Price_log'].mean().sort_values(ascending=False).head(10).plot.bar(ax=axarr[2][1], fontsize=12)
axarr[2][1].set_title("Model Vs Price", fontsize=18)

#Seventh plot: Seats Vs Price
data.groupby('Seats')['Price_log'].mean().sort_values(ascending=False).plot.bar(ax=axarr[3][0], fontsize=12)
axarr[3][0].set_title("Seats Vs Price", fontsize=18)

#Eighth Plot: Car_Age Vs Price
data.groupby('Car_Age')['Price_log'].mean().sort_values(ascending=False).plot.bar(ax=axarr[3][1], fontsize=12)
axarr[3][1].set_title("Car_Age Vs Price", fontsize=18)


plt.subplots_adjust(hspace=1.0)
plt.subplots_adjust(wspace=.5)
sns.despine()
```

**Observation** :
1) Car prices are higher in Coimbatore compared to Kolkata and Jaipur.

2) Automatic cars are more expensive than manual cars.

3) Diesel and Electric cars have similar prices, which are the highest among fuel types, while LPG cars have the lowest prices.

4) First-owner cars have the highest prices, followed by second-owner cars. Third-owner cars are less expensive than fourth and above.

5) Lamborghini is the brand with the highest prices, and the Gallardo coupe model tops the price chart.

6) 2-seater cars have the highest prices, followed by 7-seater cars.

7) The latest model cars tend to be more expensive.
